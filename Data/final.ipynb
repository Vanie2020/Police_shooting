```python
import pandas as pd
from sqlalchemy import create_engine
```

### Stored CSV files for each dataset

```python
#Store fatal shooting Set 1 CSV into DataBase

csv_file = "../Data/fatal-police-shootings-data.csv"
shooting1_df = pd.read_csv(csv_file)
shooting1_df.head()

#Store Data for 2015 set

csv_file2 = "../Data/2015NU.csv"
shooting2_df = pd.read_csv(csv_file2)

```

### Date Reformatting 

```python
shooting2_df['MonthNumber'] = shooting2_df['month']
#df

shooting2_df['MonthNumber'] = shooting2_df['MonthNumber'].replace({'January': '1', 'February': '2', 'March': '3', 'April':'4', 'May': '5', 'June': '6', 
                                                                  'July': '7', 'August': '8', 'September': '9', 'October': '10', 'November': '11', 'December': '12'})


shooting2_df['dateInt']=shooting2_df['year'].astype(str) + shooting2_df['MonthNumber'].astype(str).str.zfill(2)+ shooting2_df['day'].astype(str).str.zfill(2)
shooting2_df['NewDate'] = pd.to_datetime(shooting2_df['dateInt'], format='%Y%m%d')
shooting2_df.head()

```

### Removed Unknown for the purpose of furture accurate analyis 

```python
shooting2_df["age"].unique()
shooting2_df.drop(shooting2_df.index[shooting2_df['age'] == 'Unknown'], inplace = True)
```

```python
shooting2_df["age"].unique()
```

### Changed type from object to Datetime64

```python
#Change date from object to datetime64
shooting1_df['date'] = pd.to_datetime(shooting1_df['date'])
```

```python
shooting1_df.dtypes
```

### Merge two Data Sources to get additional columns

```python


shooting_merge = pd.merge(shooting1_df, shooting2_df, on="name", how="inner")
shooting_merge
```

### Dropping Unneeded Columns

```python
shooting_merge.drop(columns=['longitude', 'year', 'date', 'latitude', 'MonthNumber', 'city_y', 'dateInt', 'state_y', 'age_x', 'gender_x', 'race', 'flee','armed_y', 'month', 'day', 'streetaddress', 'uid', 'id' ], inplace = True)
```

```python
shooting_merge
```

### Merge to .CSV

```python
shooting_merge.to_csv(r'../Data/lastshooting.csv', index = False)
```

### Column Name Change


```python
shooting_merge.rename(columns={'armed_x': 'Armed', 'city_x': 'City', 'state_x': 'State', 'signs_of_mental_illness':'Mental Illness', 'age_y': 'Age',
                              'gender_y':'Gender', 'manner_of_death':'Manner of Death', 'threat_level':'Threat Level', 'body_camera':'Body Cam',
                              'name': 'Name', 'raceethnicity':'Race', 'classification': 'Classification', 'lawenforcementagency': 'Department', 'NewDate':'Date'})
```

### Create secured connection to local database

```python
#Connect to local dtabase

rds_connection_string = "postgres:postgres@localhost:5432/shooting_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
#check for tables 

engine.table_names()
```

### Import DF to SQL

```python
shooting_merge.to_sql(name='shooting3_db', con=engine, if_exists='append', index=False)
```
